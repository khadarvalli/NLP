## Architecture Of SpaCy.

SpaCy having the main to in It's Architures that is the 
- 1:- Document.
- 2:-  vocabulary

- The central data structures in spaCy are the Doc and the Vocab. The Doc object owns the sequence of tokens and all their annotations. The Vocab object owns a set of look-up tables that make common information available across documents. - **Serialization :-** Saving the object to file or byte strings.

# Tokenization 

Tokenization is the task of spiliting a text into the meaningful segments,called tokens.The input to be the tokenizer is a unicode text,and the output is a Doc Object.

e.g.Ishu ate the apple.

- In the above sentence the tokenization can be done as `Ishu`,`ate`,`the`,`apple.`
- In this Tokenization,It will first give attention to the spaces and at the last it focus of the `!` `,` `.` and other special chareters.

```python
nlp.remove_pipe('set_rule')
```

```python
nlp.add_pipe(set_rule,before='parser')
```

```python
def set_rule(doc):
    for token in doc[:-1]:
        if token.text =='.....':
            doc[token.i +1].is_sent_start=True
    return doc
```
